```python
import ast 

def convert_genres(text):
    l = []
    for i in ast.literal_eval(text):
        l.append(i['name'])

    return l
```

```python
major['genres'] = major['genres'].apply(convert_genres)
```

```python
major.head(2)
```

```python
import ast 

def convert_keywords(text):
    l = []
    for i in ast.literal_eval(text):
        l.append(i['name'])

    return l

major['keywords'] = major['keywords'].apply(convert_keywords)
```

```python
major.head(2)
```

```python
def convert_cast(text):
    l = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            l.append(i['name'])
        counter +=1
    return l
```

```python
major['cast'] = major['cast'].apply(convert_cast)
major.head(2)
```

```python
def convert_crew(text):
    l = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            l.append(i['name'])
        counter +=1
    return l

major['crew'] = major['crew'].apply(convert_crew)
major.head(2)
```

```python
major['overview'] = major['overview'].apply(lambda x:x.split())
major.head()
```

```python
def remove_space(word):
    l =[]
    for i in word:
        l.append(i.replace(" ",""))
    return l
```

```python
major.head()
```

```python
major['cast'] = major['cast'].apply(remove_space)
major['crew'] = major['crew'].apply(remove_space)
major['genres'] = major['genres'].apply(remove_space)
major['keywords'] = major['keywords'].apply(remove_space)
major['overview'] = major['overview'].apply(remove_space)
```

```python
ps = PorterStemmer()
```

```python
def stems(text):
    l = []
    for  i in text.split():
        l.append(ps.stem(i))
    return " ".join(l)
```

```python
print("Available Years: ", ",".join(map(str, available_years)))
selected_year = int(input("Select a year from the list above: "))

movies_by_year = new_df[new_df['year'] == selected_year]

print("\nTop 5 Movies from the Year", selected_year)
top_5_movies_year = movies_by_year[['title', 'vote_average']].sort_values(by='vote_average', ascending=False).head(5)
print(top_5_movies_year)

available_genres = new_df['genres'].unique()
print("Available Genres: ", available_genres)

selected_genre = input("\nEnter the genre of the movie: ")
movies_by_genre = new_df[new_df['genres'].str.lower().str.contains(selected_genre.lower(), case=False, na=False)]


print("\nTop Movies in the Genre", selected_genre)
top_5_movies_genre = movies_by_genre[['title', 'vote_average']].sort_values(by='vote_average', ascending=False).head(5)
print(top_5_movies_genre)

movies_choosed = input("\nEnter the movie Name: ")
movies_choosen = new_df[new_df['title'].str.lower().str.contains(movies_choosed.lower(), case=False, na=False)]
print(movies_choosen)

new_df['combined_features'] = new_df['title'] + " " + new_df['tags'] + " " + new_df['cast'] + " " + new_df['year'].astype(str) + " " + new_df['genres']

cv = CountVectorizer()
count_matrix = cv.fit_transform(new_df['combined_features'])

similarity = cosine_similarity(count_matrix)

def recommend(movie_name):
    if movie_name in new_df['title'].values:
        index = new_df[new_df['title'] == movie_name].index[0]

        selected_movie_year = new_df.loc[index,str('year')]

        distances = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda x: x[1])

        print(f"\nMovies similar to '{movie_name}' from the year {selected_movie_year}:")

        count = 0
        for i in distances[1:]: 
            movie_index = i[0]
            movie_year = new_df.loc[movie_index,str('year')]

            if movie_year == selected_movie_year:
                print(new_df.loc[movie_index, 'title'])
                count += 1
            if count ==6:
                break
    else:
        print(f"Movie '{movie_name}' is not found on our dataset.")


```
